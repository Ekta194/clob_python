```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with three separate vectors for the horizontal axis
data1 = pd.DataFrame({
    'X1': [1, 2, 3, 4, 5, 6],
    'Y': [10, 15, 12, 17, 13, 16]
})

data2 = pd.DataFrame({
    'X2': [2, 4, 6, 8, 10, 12],
    'Y': [11, 17, 14, 12, 19, 10]
})


data3 = pd.DataFrame({
    'X3': [3, 6, 9, 12, 15, 18],
    'Y': [17, 13, 18, 12, 12, 18]
})




# Plotting the scatter plot
plt.scatter(data1['X1'], data1['Y'], color='red', label='Trend 1')
plt.scatter(data2['X2'], data2['Y'], color='green', label='Trend 2')
plt.scatter(data3['X3'], data3['Y'], color='blue', label='Trend 3')

# Customize the plot
plt.xlabel('Horizontal Axis')
plt.ylabel('Vertical Axis')
plt.title('Scatter Plot with Multiple Trends')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
category_amount = {
    'Category A': 15,
    'Category B': 12,
    'Category C': 8,
    'Category D': 10,
    'Category E': 5
}

# Extract category names and corresponding amounts
categories = list(category_amount.keys())
amounts = list(category_amount.values())

# Plotting the bar graph
plt.bar(categories, amounts)

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Amount')
plt.title('Bar Graph')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
